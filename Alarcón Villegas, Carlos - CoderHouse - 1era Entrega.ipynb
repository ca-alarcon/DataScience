### **1. Predicción de un derrame cerebral.**

Un **derrame cerebral** ocurre cuando se interrumpe el flujo sanguíneo al cerebro. Pocos minutos después, que se produce un derrame cerebral, las **células del cerebro comienzan a morir** lo que puede causar daños a largo plazo. 

<br>

**Algunos números de la Organización Mundial de la Salud:**
* En el mundo, 1 de cada 6 personas tendrá un derrame cerebral en su vida. 
* Cada año, 15 millones de personas lo experimentan y 6 millones de personas no sobreviven. 
* La mayoría de personas que padecen un derrame cerebral tendrán una discapacidad residual.

<br>

Los factores de riesgo son comportamientos o rasgos que hacen más propenso a desarrollar una enfermedad o afección. Muchos de los factores de riesgo pueden ser tratados, modificados o controlados. Sin embargo, algunos factores de reisgo para el derrame cerebral no pueden ser cambiados. 

<br>

El factor principal de riesgo relacionado con el derrame cerebral que puede ser tratado es la presión arterial. En la tabla, corresponde a la columna **"hypertension"**. La presión arterial alta no causa signos ni síntomas, por eso es importante tomarsela regularmente. 

<br>

Para el análisis, también se tendrá en cuenta si el individuo cuenta con diabetes. La diabetes aumenta el riesgo de accidente, especialmente para derrames cerebrales debido al daño de pequeños vasos sanguíneos. En este caso, pondremos atención en la columna **"avg_glucose_level"** y contemplaremos tres escenarios de acuerdo con el nivel de azúcar en la sangre: 
1. Valores menores a 99 mg/dl son normales.
2. Valores entre 100 a 125 mg/dl indican que el individuo tiene pre-diabetes.
3. Valores mayores a 126 mg/dl indican que tiene diabetes. 

<br>

Contemplaremos la edad de las personas, correspondiente a la columna **"age"**, ya que individuos de 55 años o mayores corren un riesgo más alto de sufrir un accidente cerebrovascular que las personas más jóvenes.

### **2. Predicción de aprobación de tarjeta de crédito.**

Una entidad bancaria evalúa varios factores al momento de aprobar una solicitud de tarjeta de crédito, ya que deben determinar si los solicitantes son capaces de cumplir con sus obligaciones de pago. Esto les permite minimizar el riesgo de impagos y morosidad, y también les permite determinar el límite de crédito (Disponible y Disponible en cuotas) que se les puede otorgar. 

<br>

A continuación, se realizará un modelo para indicar si un solicitante es "bueno" o "malo", para ello, he definido las siguientes variables como potencialmente valiosas a tener en cuenta: 

<br>

* **Ingresos:** Indican la capacidad de una persona de generar recursos para cubrir sus gastos, incluyendo el pago de sus deudas. Para ello, pondremos atención en la columna **"AMT_INCOME_TOTAL"**, correspondiente a los ingresos anuales del solicitante. 

<br>

* **Historial crediticio:** Les proporciona información sobre el comportamiento crediticio del solicitante en el pasado. Un historial crediticio positivo, es decir, con pagos puntuales (en fecha), indica que el solicitante es una persona responsable y confiable en cuanto a sus obligaciones de pago. En cambio, un historial crediticio negativo, es decir, con pagos tardíos o impagos, indica que el solicitante tiene dificultades para cumplir con sus obligaciones de pago y es más propenso a tener problemas de impagos. 
En este caso, contemplaremos la columna **"STATUS"** que nos entrega el retraso en el pago por parte del solitante. Se cuenta con los siguientes escenarios: 

1. "0": 1-29 días de retraso en el pago
2. "1": 30-59 días de retraso en el pago
3. "2": 60-89 días de retraso en el pago
4. "3": 90-119 días de retraso en el pago
5. "4": 120-149 días de retraso en el pago
6. "5": Deudas vencidas o malas, escrituras de más de 150 días
7. "C": pagado ese mes
8. "X": No hay préstamo para ese mes

<br>

* **Antigüedad:** Proporciona información sobre su estabilidad laboral. Si un trabajador tiene una antigüedad prolongada en una empresa, es probable que tenga un empleo estable y una fuente de ingresos confiable. Podría indicar que es más probable que sea capaz de cumplir con sus obligaciones de pago en el futuro. Para ello, contemplaremos la columna **"DAYS_EMPLOYED"** que hace referencia a los días en que el solicitante se encuentra trabajando (valores negativos) o desempleado (valor positivo).
